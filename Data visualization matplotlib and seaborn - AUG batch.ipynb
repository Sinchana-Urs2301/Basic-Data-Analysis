# Axes, Labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t = [52,51,50,48,49,46,47]
min_t = [36,42,40,44,43,35,35]
avg_t = [48,46,45,46,46,40,41]
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weater forecast")
plt.plot(days, max_t, label = "MAX")
plt.plot(days, min_t, label = "MIN")
plt.plot(days, avg_t, label = "AVG")
plt.legend(loc ="lower center", shadow = False , fontsize = "large")
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weater forecast")
plt.plot(days, max_t, label = "MAX")
plt.plot(days, min_t, label = "MIN")
plt.plot(days, avg_t, label = "AVG")
plt.legend(loc ="best", shadow = True , fontsize = "large")
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weater forecast")
plt.plot(days, max_t, label = "MAX")
plt.plot(days,min_t, label = "MIN")
plt.plot(days,avg_t, label = "AVG")
plt.legend(loc ="best", shadow = False, fontsize = "large")
plt.grid()
```

# BAR GRAPH

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
```

```python
plt.bar(xpos, revenue)
```

```python
plt.bar(xpos, revenue, label = "Revenue")
plt.xticks(xpos, company)
plt.ylabel("Revenue(bln)")
plt.title("US TECH STOCKS")
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos, company)
plt.ylabel("rev in bln")
plt.title("US Tech Stock")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos, profit, label = "Profit")
plt.legend()
```

```python
plt.bar(xpos - 0.2, revenue, width = 0.4, label = "Revenue")
plt.bar(xpos +0.2 , profit, width = 0.4, label= "Profit")
plt.xticks(xpos, company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Horizontal Bar Graph

```python
plt.barh(xpos, revenue, label = "REV")
plt.yticks(xpos, company)
plt.xlabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2, revenue,height = 0.4, label = "Revenue" )
plt.barh(xpos+0.2, profit, height =0.4, label = "Profit")
plt.yticks(xpos,company)
plt.xlabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# HISTOGRAM

- In histograms, x axis contains a variable and y axis will be a frequency of that variable We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, rwidth = 0.8)
```

```python
plt.hist(blood_sugar, rwidth = 0.8, bins = 4)
```

```python
plt.hist(blood_sugar, rwidth = 0.8, bins = [80,100,125,150], color = "g")
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

# Mutiple data samples in a histogram

```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women], rwidth = 0.95, bins = [80,100,125,150], color = ["green", "orange"], label =["Men","Women"])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype = "step")
```

# Horizontal histogram
### orientation

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95,orientation  = "horizontal")
```
