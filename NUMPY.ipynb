# Concatinate

```python
a1 = [1,2,3,4,5]
a2 = [6,7,8,9,10]
aa = np.array(a1)
ab = np.array(a2)
print(aa)
print(ab)
```

```python
v = np.vstack((aa,ab))
v
```

```python
h = np.hstack((aa,ab))
h
```

```python
a10 = [1,2,3,4,5,6]
a11 = [7,8,9,10,11,12]
ar = np.array(a10)
br = np.array(a11)
arr =ar.reshape((2,3))
brr =br.reshape((2,3))
print(arr)
print(brr)
```

```python
ve = np.vstack((arr,brr))
ve
```

```python
ho = np.hstack((arr,brr))
ho
```
