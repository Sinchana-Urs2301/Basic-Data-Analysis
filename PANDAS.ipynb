# CONCATINATE
# Append the datasets

```python
df7 = pd.DataFrame({"customer":['101','102','103','104'], 
                    'category': ['cat2','cat2','cat1','cat3'],
                    'important': ['yes','no','yes','yes'],
                    'sales': [123,52,214,663]},index=[0,1,2,3])

df8 = pd.DataFrame({"customer":['101','103','104','105'], 
                    'color': ['yellow','green','green','blue'],
                    'distance': [12,9,44,21],
                    'sales': [123,214,663,331]},index=[4,5,6,7])
```

```python
print(df7)
print(df8)
```

```python
pd.concat([df7,df8],axis = 1, sort = True)
```

```python
pd.concat([df7,df8], axis = 0, sort= False)
```

# JOIN 
# COMBINE TWO OR MORE DATASETS BASED ON INDEX

```python
df9 = pd.DataFrame({'Q1': [101,102,103],
                    'Q2': [201,202,203]},
                   index=['I6','I4','I5'])

df10 = pd.DataFrame({'Q3': [301,302,303],
                    'Q4': [401,402,403]},
                   index=['I0','I2','I3'])

print(df9)
print()
print(df10)
```

```python
df9.join(df10 , how="inner")
```

```python
df9.join(df10, how="outer")
```

```python
df9.join(df10, how="left")
```

```python
df10.join(df9, how="left")
```

```python
df1 = pd.DataFrame({'Q1': [101,102,103],
                    'Q2': [201,202,203]},
                   index=['I0','I1','I2'])

df2 = pd.DataFrame({'Q3': [301,302,303],
                    'Q4': [401,402,403]},
                   index=['I0','I2','I3'])

print(df1)
print()
print(df2)
```

```python
df1.join(df2, how="inner")
```

```python
df1.join(df2, how="left")
```

```python
df2.join(df1, how="left")
```

```python
df1.join(df2, how="outer")
```

# GROUPBY

```python
da  =  {'items' : ['veg_burger', 'veg_burger', 'chicken_burger','chicken_burger', 'softy_icecream', 'softy_icecream'], 
       
       'sale' : [100, 200, 100, 250, 150, 50] , 'cities' : ['c1', 'c2', 'c1', 'c2', 'c1', 'c2']}

sales  =  pd.DataFrame(da)
sales
```

```python
sales.groupby(by = "items").max()
```

```python
sales.groupby(by = "sale").max()
```

```python
sales.groupby(by = "sale").min()
```

```python
sales.groupby(by = "sale").count()
```
